```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x=np.arange(0,10,0.1)
y=2*x+1
```

```python
plt.plot(x,y,linewidth=2.0,linestyle='--',color='orange',alpha=0.5,marker='o')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("y=2x+1")
```

```python
#normally graph
plt.plot(x,y)
plt.show()
```
